# Histogram Pokemon
Hieronder zal een pokemon gerelateerde histogram geplot worden.

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("pokemon_small.csv")

plt.figure(figsize=(10, 10))
sns.histplot(df["HP"])
plt.show()

plt.figure(figsize=(10, 10))
sns.histplot(df["HP"], bins=35)
plt.show()
```
