# Histogram Pokemon
Hieronder zal een pokemon gerelateerde histogram geplot worden.

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("pokemon_small.csv")

plt.figure(figsize=(10, 10))
sns.histplot(df["HP"])
plt.show()

plt.figure(figsize=(10, 10))
sns.histplot(df["HP"], bins=35)
plt.show()
```

## Scatterplot Pokemon
Hieronder komt de pokemon gerelateerde scatterplot grafiek te staan.

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(data=df, x='Attack', y='HP')
plt.show()

plt.figure(figsize=(8, 8))
sns.scatterplot(data=df, x='Attack', y='Speed')
plt.show()

plt.figure(figsize=(8, 8))
sns.scatterplot(data=df, x='Attack', y='HP', hue='Stage', palette='magma')
plt.show()

plt.figure(figsize=(8, 8))
sns.scatterplot(data=df, x='Attack', y='Speed', hue='Stage', palette='magma')
plt.show()
```

### Boxplot Pokemon
Hieronder komen de pokemon gerelateerde boxplot grafieken te staan.

```python
plt.figure(figsize=(8, 8))
sns.boxplot(data=df, x="Stage", y="Total")
plt.show()

plt.figure(figsize=(8, 8))
sns.boxplot(data=df, x="Stage", y="Total", hue="Legendary")
plt.show()
```

#### Violinplot Pokemon
Hieronder komen de pokemon gerelateerde violinplot grafieken te staan.

```python
filtered_df = df[(df["Type 1"].isin(["Water", "Grass", "Fire", "Bug"])) & (df["Stage"] < 3)]
plt.figure(figsize=(8, 8))
sns.violinplot(data=filtered_df, x="Type 1", y="Total")
plt.show()

plt.figure(figsize=(8, 8))
sns.violinplot(data=filtered_df, x="Type 1", y="Total", hue="Stage", split=True)
plt.show()

plt.figure(figsize=(8, 8))
sns.violinplot(data=filtered_df, x="Type 1", y="Total", hue="Stage", split=True, scale="count")
plt.show()

plt.figure(figsize=(8, 8))
sns.violinplot(data=filtered_df, x="Type 1", y="Total", hue="Stage", split=True, scale="count", cut=0)
sns.swarmplot(data=filtered_df, x="Type 1", y="Total", hue="Stage", dodge=True, palette=["black"], alpha=0.7)
plt.show()
```
