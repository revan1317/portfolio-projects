# Lijnplot Pokemon
Hieronder komen de pokemon gerelateerde lijngrafieken te staan.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("pokemon_extended.csv", index_col="name")

df.groupby("generation")["weight_kg"].mean().plot()
plt.ylabel("weight_kg")
plt.show()

df.groupby("generation")["height_m"].mean().plot()
plt.ylabel("height_m")
plt.show()
```

## Piechart Pokemon
Hieronder komt de pokemon gerelateerde piechart te staan.

```python
type_counts = df["type1"].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index)
plt.axis('equal')
plt.show()
```

### Styling
Hieronder staat de pokemon gerelateerde, gestylde staafgrafiek.


```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

sns.countplot(data=df, x="type1", ax=ax[0], order=df["type1"].value_counts().index, palette="husl")
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=45)

sns.countplot(data=df, x="type2", ax=ax[1], order=df["type2"].value_counts().index, palette="husl")
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```
