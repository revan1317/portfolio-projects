# Lijnplot Pokemon
Hieronder komen de pokemon gerelateerde lijngrafieken te staan.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("pokemon_extended.csv", index_col="name")

df.groupby("generation")["weight_kg"].mean().plot()
plt.ylabel("weight_kg")
plt.show()

df.groupby("generation")["height_m"].mean().plot()
plt.ylabel("height_m")
plt.show()
```
