# Controleren van vetwaarden in smoothies

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("mc_menu_final.csv")

smoothies_shakes = df[df['Category'] == 'Smoothies & Shakes']

plt.figure(figsize=(10, 6))
plt.bar(smoothies_shakes['Item'], smoothies_shakes['Total Fat'], label='Total Fat')
plt.xticks(rotation=90)
plt.xlabel('Item')
plt.ylabel('Total Fat')
plt.legend()
plt.tight_layout()
plt.show()
```

## Foutieve vetwaarden

In de grafiek hierboven zie je dat sommige smoothies een negatieve waarde hebben voor het aantal gram vet. Dat klopt natuurlijk niet want vet kan niet negatief zijn.

De volgende items hebben een foute waarde:
- Mango Pineapple Smoothie (Small)
- Mango Pineapple Smoothie (Medium)
- Mango Pineapple Smoothie (Large)
- Strawberry Banana Smoothie (Small)
- Strawberry Banana Smoothie (Medium)
- Strawberry Banana Smoothie (Large)
