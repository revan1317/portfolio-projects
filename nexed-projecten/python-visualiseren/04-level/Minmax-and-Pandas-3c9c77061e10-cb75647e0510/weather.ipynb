### Specifiek Amsterdam

Hieronder wordt alleen de gemiddelde temperatuur van Amsterdam ingezien.

```python
amsterdam_df = weather_nl[weather_nl["City"] == "Amsterdam"]

amsterdam_df.plot(x="dt", y="AverageTemperature", title="Gemiddelde temperatuur in Amsterdam (1960â€“2013)", figsize=(12, 6))
plt.xlabel("Datum")
plt.ylabel("Gemiddelde temperatuur (Â°C)")
plt.grid(True)
plt.show()

```

#### Hoogste gemiddelde Rotterdam
Hieronder komt de hoogste gemiddelde temperatuur van Rotterdam te staan.

```python
max_temp_rotterdam = weather_nl[weather_nl["City"] == "Rotterdam"]["AverageTemperature"].max()
print(f"Maximale temperatuur: {max_temp_rotterdam:.2f} ")

```

##### Laagste gemiddelde Maastricht
Hieronder komt de laagste gemiddelde temperatuur van Maastricht te staan.

```python
min_temp_maastricht = weather_nl[weather_nl["City"] == "Maastricht"]["AverageTemperature"].min()
print(f"Minimale temperatuur: {min_temp_maastricht:.2f} ")
```

Aantal steden

```python
aantal_steden = weather_nl["City"].nunique()
print(f"Aantal steden: {aantal_steden}")
```

Meest voorkomende temperatuur

```python
meest_voorkomende_temp = weather_nl["AverageTemperature"].dropna().round().mode()[0]
print(f"Meest voorkomende temperatuur: {meest_voorkomende_temp} ")
```

3 warmste temperaturen

```python
warmste_temperaturen = weather_nl["AverageTemperature"].dropna().nlargest(3)

for temp in warmste_temperaturen:
    print(f"{temp:.2f} ")

```
