# Uitleg
Dit is de top 10 van de clubs met de hoogste totale spelerswaarde getoond in een barchart.

```python
import pandas as pd, matplotlib.pyplot as plt

df = pd.read_csv("fifa_ratings.csv")
top10 = df.groupby("club")["value_eur"].sum().sort_values(ascending=False).head(10)

top10.plot(kind='bar', figsize=(8,6), legend=True)
plt.xticks(rotation=90)
plt.xlabel("club")
plt.ylabel("value_eur")
plt.tight_layout()
plt.show()
```
