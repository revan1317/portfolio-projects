## Heeft iedereen gestemd?

We controleren of iedereen heeft gestemd! Als het goed is, dan zijn er 51 staten in totaal. Dit zijn de 50 staten van de Verenigde Staten plus Washington D.C, wat in deze database als een staat wordt beschouwd.

```python
df = pd.read_csv('president_county_candidate.csv')
unieke_staten = df['state'].nunique()
print(f"Aantal unieke staten in de dataset: {unieke_staten}")

if unieke_staten == 51:
    print("Alle staten hebben hun stemmen opgestuurd!")
else:
    print("Niet alle staten zijn aanwezig in de dataset.")
```

Stem teller hier onder

```python
trump_stemmen = df[df['kandidaat'] == 'Donald Trump']['stemmen'].values[0]
biden_stemmen = df[df['kandidaat'] == 'Joe Biden']['stemmen'].values[0]
andere_stemmen = df[
    (df['kandidaat'] != 'Donald Trump') & 
    (df['kandidaat'] != 'Joe Biden')
]['stemmen'].sum()

aangepaste_trump_stemmen = trump_stemmen + andere_stemmen

if aangepaste_trump_stemmen > biden_stemmen:
    winnaar = "Donald Trump"
else:
    winnaar = "Joe Biden"

print(f"Stemmen Joe Biden: {biden_stemmen}")
print(f"Stemmen Donald Trump (met extra stemmen): {aangepaste_trump_stemmen}")
print(f"Winnaar volgens Trump: {winnaar}")
```
