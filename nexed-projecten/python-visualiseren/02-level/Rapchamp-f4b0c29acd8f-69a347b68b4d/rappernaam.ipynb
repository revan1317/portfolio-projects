# Maak je eigen rappernaam
Dit is een document waarin we een rappernaam maken.

```python
import random
```

In de bovenstaande code hebben we de random module ge√Ømporteerd. Hieronder maken we wat variabelen aan.

```python
naam = input("Wat is jouw naam?")
meubelstuk = input("Wat is je favortiete meubelstuk")
keukengerei = input("Welk voorwerp uit de keuken heb je als laatst gebruikt")
voedsel = input("Wat is je lievelingseten")
kleur = input("Wat is je lievelingskleur")
```


In het volgende stukje code wordt er een functie gedefinieerd die een rappernaam kan maken. Deze functie gaan we later gebruiken om voor de gebruiker een rappernaam te bouwen.

```python
def genereer_rapper_naam(naam, kleur, gek_woord):
    voorstukje = random.choice(["Lil", "Big", kleur, "Mac", "MC"])

    if random.random() < 0.25:
        return voorstukje + ' "' + naam + '" ' + gek_woord
    else:
        return voorstukje + ' ' + gek_woord
```

Als je het volgende stukje code runt, dan genereer je je eigen rappernaam! Als je een nieuwe naam wil, kun je het stukje nog een keer runnen. Als je andere variabelen wil gebruiken, dan kun je een nieuwe naam, leeftijd en lievelingskleur bovenaan opnieuw invullen.

```python
gek_woord = random.choice([meubelstuk, keukengerei, voedsel])

rappernaam = genereer_rapper_naam(naam, kleur, gek_woord)

print("Jouw rappernaam is " + rappernaam)
```
